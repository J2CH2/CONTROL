```python
import matplotlib.pyplot as plt
import control as co
import numpy as np
from sympy import *
R=18000
C=0.000470
num=[0,1] 
num=np.array(num) 
den=[R*C,1] 
den=np.array(den) 
print(num)
print(den)
```

```python
sys=co.tf(num,den)
print(sys)
```

```python
x=co.poles(sys)
print(x)
#plt.figure(1)
#gain=co.root_locus(sys,kvect=np.linspace(-4,1,5))
l=co.root_locus(sys,kvect=np.linspace(-0.7,-0.6,10),xlim=(-1,1),ylim=(-1,1),grid=True)
display(l)
```

```python
t=np.linspace(0,40,100)
t1,y1=co.step_response(sys,t)
plt.plot(t1,y1,'r--')
plt.xlabel('time(s)')
plt.ylabel('amplitud')
plt.grid()
plt.show()
```
